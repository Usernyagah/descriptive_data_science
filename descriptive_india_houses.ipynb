In this assignment, you'll work with a dataset of homes for sale in India. Your goal is to determine if there are regional differences in the real estate market.

```python
# Import Matplotlib, pandas
import matplotlib.pyplot as plt
import pandas as pd
```

Prepare Data

Import first dataset and clean the data

```python
df1 = pd.read_csv("data/Bangalore.csv")
df1.head()
```

```python
df1.info()
```

Drop all rows with NaN values from the DataFrame df1

```python
df1.dropna(inplace=True)
```

Drop the unnecessary columns from df1

```python
df1 = df1[['Price', 'Area', 'Location']]
```

```python
df1.head()
```

```python
df1.info()
```

Import dataset 2 and clean it up

```python
df2 = pd.read_csv("data/Chennai.csv")
df2.head()
```

Cleaning it up

```python
df2.dropna(inplace=True)
df2 = df2[['Price', 'Area', 'Location']]
df2.head()
```

Import dataset 3 

```python
df3 = pd.read_csv("data/Delhi.csv")
df3.head()
```

Cleaning the dataset

```python
df3.dropna(inplace=True)
df3 = df3[['Price', 'Area', 'Location']]
df3.head()
```

Importing dataset 4 

```python
df4 = pd.read_csv("data/Hyderabad.csv")
df4.head()
```

Cleaning it up

```python
df4.dropna(inplace=True)
df4 = df4[['Price', 'Area', 'Location']]
df4.head(10)
```

```python
df4.info()
```

Importing dataset 5

```python
df5 = pd.read_csv("data/Kolkata.csv")
df5.head()
```

Cleaning it up

```python
df5.dropna(inplace=True)
df5 = df5[['Price', 'Area', 'Location']]
df5.head(10)
```

importing the last dataset

```python
df6 = pd.read_csv("data/Mumbai.csv")
df6.head(1111)
```

Cleaning the dataset

```python
df6.dropna(inplace=True)
df6 = df6[['Price', 'Area', 'Location']]
df6.head(10)
```

Concatenate df1,df2,df3,df4,df4 and df6 to create a new DataFrame named df

```python
df = pd.concat([df1,df2,df3,df4,df5,df6])
df.head()
```

```python
df.info()
```

Explore the clean data

 Use the describe method to create a DataFrame summary_stats with the summary statistics for the "Area" and "Price" columns

```python
summary_stats = df[["Area","Price"]].describe()
summary_stats
```

Create a histogram of "Price"

```python
low, high = df["Area"].quantile([0.1, 0.9])
df = df[df["Area"].between(low, high)]
df["Area"].describe()
```

```python
low, high = df["Price"].quantile([0.1, 0.9])
df = df[df["Price"].between(low, high)]
df["Price"].describe()
```

```python
# Build histogram
plt.hist(df["Price"])


# Label axes
plt.xlabel("Price [ndian Rupee]")
plt.ylabel("Frequency")
# Add title
plt.title("Distribution of Home Prices");
```

Create a horizontal boxplot of "Area"

```python
# Build box plot
plt.boxplot(df["Area"] ,vert=False)


# Label x-axis
plt.xlabel("Area [sq meters]")

# Add title
plt.title("Distribution of Home Sizes")
```

Use the groupby method to create a Series named mean_price_by_region that shows the mean home price in each region in India, sorted from smallest to largest.

```python
mean_price_by_region = df.groupby("Location")["Price"].mean().sort_values().head(10)
mean_price_by_region
```

Use mean_price_by_region to create a bar chart

```python
mean_price_by_region.plot(
    kind="bar",
    xlabel="Location",
    ylabel="Mean Price [Indian Rupees]",
    title="Mean Home Price by Region"

);
```

Using the corr method, calculate the Pearson correlation coefficient for "area" and "price"

```python
correlation = df["Area"].corr(df["Price"])
print("Correlation of 'Area' and 'Price' (all India):", correlation)
```

Create a scatter plot of "price" vs "Area"

```python
plt.scatter(x=df['Area'], y=df["Price"])

# Add x-axis label

plt.xlabel("Area ")
# Add y-axis label
plt.ylabel("Price [Indian Rupees]")

# Add title
plt.title("India: Price vs. Area");
```

Conclusion

From the data there is regional differences in real estate markets
